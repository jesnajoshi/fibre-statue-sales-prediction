```python
df["Date"]=pd.to_datetime(df["Invoice_Date"])
df.set_index('Invoice_Date')['Total_Sales'].plot(figsize=(15,5))
```

```python
topsellingstatue=df.groupby('Statue_Name')['Total_Sales'].sum().sort_values(ascending=False)
```

```python
print("Total_Sales by statue:")
print(topsellingstatue)
```

```python
category_sales=df.groupby('Category')['Total_Sales'].sum().reset_index()
print("Total_Sales by Category:")
print(category_sales)

plt.figure(figsize=(5,4))
sns.barplot(data=category_sales,x='Category',y='Total_Sales',palette='inferno',hue='Category')
plt.title('Total Sales by Category:god vs Animal Statues')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.show()
```

```python
df['Category'].unique()
```

```python
statue_performance=df.groupby('Statue_Name')['Quantity_Sold'].sum().reset_index()
plt.figure(figsize=(5,3))
sns.barplot(data=statue_performance,x='Quantity_Sold',y='Statue_Name',palette='coolwarm',hue='Quantity_Sold')
plt.title('Total Selling statue by quantity')
plt.xlabel('Quantity_Sold')
plt.ylabel('Statue_Name')
plt.show()
```

```python
#profitablity by statue size
```

```python
# 1. Group by Size and calculate Average Price and Total Sales
size_analysis = df.groupby('Size').agg({'Price': 'mean', 'Total_Sales': 'sum'}).reset_index()

# 2. Plotting the Revenue by Size
plt.figure(figsize=(5,3))
sns.barplot(data=size_analysis, x='Size', y='Total_Sales', palette='magma',hue='Size' ,order=['Small', 'Medium', 'Large'])
plt.title('Revenue Contribution by Statue Size')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
category_sales=df.groupby("Category")["Sales"].sum()
plt.figure()
category_sales.plot(kind="bar")
plt.title("Category- wise total sales")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.show()
```

```python
top_animal = (
    df.groupby( 'Statue_Name')['Total_Sales']
      .sum()
      .reset_index()
      .sort_values( 'Total_Sales', ascending=False)
      .head(1)
)
plt.figure(figsize=(5,4))
sns.barplot(data=top_animal,x='Total_Sales',y='Statue_Name',palette='inferno',hue='Category')
plt.title('Total Sales by Category:god vs Animal Statues')
plt.xlabel('Total Sales')
plt.ylabel('Statue_Name')
plt.show()

```

Sales data shows seasonal variation, with higher demand during festival and celebration periods.

Festival months contribute the maximum sales, indicating that customer purchases are event-driven.

Price has a negative correlation with sales quantity, meaning as price increases, the number of units sold decreases.

Moderately priced fibre statues achieve higher sales volume compared to high-priced statues.
Product category influences sales, where religious and traditional fibre statues have higher demand than decorative ones.

Discounts and offers positively impact sales, increasing customer purchase behavior.
Sales trend shows gradual growth over time, indicating rising popularity of fibre statues in the market.

Large-sized statues generate higher revenue, though smaller statues sell in greater quantities.

Historical sales data is a strong predictor of future sales when combined with seasonal features.
